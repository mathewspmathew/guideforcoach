```python
import pandas as pd
from pathlib import Path
import joblib
```

```python
current = Path.cwd()
df = pd.read_csv(current.parent / "data" / "players_data_light-2024_2025.csv")
```

```python
df_raw = df[df["Pos"]  != "GK"]
df_raw.shape
```

removing gk

```python
df_raw = df_raw[df_raw["90s"] > 5.0].copy()
# to remove outliers - remove players less than 5 games played
df_raw.shape
```

```python
# for compairing player with 20 goals in 5 games and player with 25 goals in 10 games are diff right
# for Attack
df_raw['npxG_p90'] = df_raw['npxG'] / df_raw['90s']

# for Creation
df_raw['xAG_p90'] = df_raw['xAG'] / df_raw['90s']
df_raw['PrgP_p90'] = df_raw['PrgP'] / df_raw['90s']
df_raw['KP_p90'] = df_raw['KP'] / df_raw['90s']

# Possession
df_raw['PrgC_p90'] = df_raw['PrgC'] / df_raw['90s']

# Defense
df_raw['TklW_p90'] = df_raw['TklW'] / df_raw['90s']
df_raw['Int_p90'] = df_raw['Int'] / df_raw['90s']
df_raw['Recov_p90'] = df_raw['Recov'] / df_raw['90s']

```

```python
df_raw.reset_index(drop=True)
print(f"Data Loaded & Cleaned. active Players: {len(df_raw)})")
```
